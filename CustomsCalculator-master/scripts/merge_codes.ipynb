```python
import pandas as pd
import re
```

```python
def clean_code(code):
	"""Удаляет пробелы и оставляет только цифры"""
	if pd.isna(code): return None
	return re.sub(r'\D', '', str(code))
```

```python
def get_parent(code):
	"""Создает родительский код для иерархии"""
	if not code: return None
	code_len = len(code)
	if code_len == 10:
		return code[:6]
	elif code_len > 6:
		return code[:6]
	elif code_len == 6:
		return code[:4]
	elif code_len == 5:
		return code[:4]
	elif code_len == 4:
		return code[:2]
	return None
```
