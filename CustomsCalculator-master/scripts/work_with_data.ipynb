```python
def normalize_unit(unit_str):
	# ... твой код ...
	if not unit_str: return None
	u = unit_str.lower().strip()
	if 'килограмм' in u or 'кг' in u: return 'kg'
	if 'литр' in u: return 'l'
	if 'штук' in u or 'шт' in u:
		if '1000' in u: return '1000_pcs'
		return 'pcs'
	if 'пар' in u: return 'pair'
	if 'куб' in u or 'см' in u: return 'cm3'
	if 'м2' in u: return 'm2'
	return u
```
